# PyWaffle Library
*[documentacion](https://pywaffle.readthedocs.io/en/latest/)*

```python
from pywaffle import Waffle
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(
    FigureClass = Waffle,
    rows        = 5,
    values      = {'One':30, 'Two':15, 'Three':5},
    legend      = {'loc':'upper left', 'bbox_to_anchor':(1,1)},
    title       = {'label':'Hello Waffles', 'loc':'center','fontdict':{'fontsize':20}},
    #colors      = ["#232066", "#983D3D", "#F5D55C"],
    cmap_name   = "Dark2",
    icons       = 'child',
    font_size   = 26,
    facecolor   = '#DDDDDD',
    #plot_anchor = 'C', #Alignment method for sublots
    interval_ratio_x = 0.5,#Icon spacing
    interval_ratio_y = 1,
    vertical         = True, #Plot direction
)
```

```python
fig = plt.figure(
    FigureClass = Waffle,
    rows        = 5,
    values      = {'Walk':30, 'Bus':15, 'Car':5},
    legend      = {'loc':'upper left', 'bbox_to_anchor':(1,1)},
    title       = {'label':'Hello Waffles', 'loc':'center','fontdict':{'fontsize':20}},
    cmap_name   = "Dark2",
    icons       = ['child', 'bus', 'car'],
    font_size   = 26,
    facecolor   = '#DDDDDD'
)
```
